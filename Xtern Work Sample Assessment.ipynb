# Import FIle

```python
df = pd.read_csv (r'C:\2020-XTern-DS.csv')
```

```python
df.head(2)
```

# Fix / Normalize Data

```python
df.shape
```

Remove any records where Average_Cost column has non-currency values

```python
df = df[df.Average_Cost.str[:1] == '$']
```

Convert Average_Cost to numeric 

```python
df ['Average_Cost'] = np.float32(df.Average_Cost.str.replace('$',''))
```

```python
df.shape
```

Convert Reviews column to numeric

```python
df['Reviews'] = df.Reviews.str.replace('-','0')
```

```python
df['Reviews'] = np.int32(df.Reviews)
```
